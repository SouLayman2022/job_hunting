```python
import matplotlib.pyplot as plt

categories = ['Étudiants en Master\n (Public)', 'Étudiants en Master\n (Privé)', 'Offres d\'alternances']
values = [579800, 250512, 400000]  

plt.figure(figsize=(10, 6))
bars = plt.bar(categories, values, color=['blue', 'orange', 'green'])

# Add labels directly to bars
for bar, value in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{value}', ha='center', va='bottom')

plt.title('Nombre d\'étudiants en Master et offres d\'alternances')
plt.xlabel('\nCategories')
plt.ylabel('Effectifs\n')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.yticks(range(0, 1000000, 100000))  

plt.legend(bars, categories)

plt.tight_layout()

plt.show()
```
